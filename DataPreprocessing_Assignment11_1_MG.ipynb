**Step 1: Importing the libraries**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder , OneHotEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split

```

**Step 2: Importing dataset**

```python
df=pd.read_csv("/content/drive/MyDrive/Guvi_collab/Data.csv")
df
```

```python
df.info()
```

**Step 3: Handling the missing data**

```python
df = df.dropna()
df
```

**Step 4: Encoding categorical data**

```python
df_1=df.copy()
le=LabelEncoder()
for column in df.columns:
    if df[column].dtype=='object':
          df_1[column]=le.fit_transform(df[column])
```

```python
df_1
```

**Step 5: Creating a dummy variable**

```python
df_2=pd.get_dummies(df,columns=['Country','Purchased'])
df_2
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
x=df.drop('Salary',axis=1)
y=df['Salary']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30)
```

```python
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

**Step 7: Feature Scaling**

```python
  #now we'll scale our data salary column and age column since it's the only variable in the data which is wide ranges
MMS = MinMaxScaler(feature_range=(0,1)) #will set the scaling range to 0 to 1.
```

```python
columns=['Salary','Age']
for column in columns:
  df_1[column]=MMS.fit_transform(df_1[[column]])
```
